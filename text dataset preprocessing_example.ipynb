```python
import konlpy.tag

with open('한화투자증권-남성현.txt', 'r', encoding='utf8') as f:
    content = f.read()
    
filtered_content = content.replace('.', '').replace(',','').replace("'","").replace('·', ' ').replace('=','').replace('\n','').replace('-','').replace('):','').replace('(','').replace('[','').replace(']','').replace(':','')

Okt = konlpy.tag.Okt()
Okt_morphs = Okt.pos(filtered_content)  # 튜플반환
print(Okt_morphs)
```

```python
# 명사만 추출
Noun_words = []
for word, pos in Okt_morphs:
    if pos == 'Noun':
        Noun_words.append(word)
print(Noun_words)
```

```python
# 불용어 제거용 사전 구축

# stopwords = ['성장']
unique_Noun_words = set(Noun_words)
for word in unique_Noun_words:
    if word in stopwords:
        while word in Noun_words: Noun_words.remove(word)  # 최종결과 : Noun_words
```

```python
# 빈도 분석
from collections import Counter

c = Counter(Noun_words)
print(c.most_common(10)) # 상위 10개 출력하기
```
